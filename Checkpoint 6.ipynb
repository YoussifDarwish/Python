```python
import pandas as pd
import numpy as np
```

```python
dict1 = {"name": ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        "score": [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        "attempts": [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        "qualify": ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

df = pd.DataFrame(dict1, index = labels)
```

```python
dict2 = {"name": "Suresh", "score": 15.5, "attempts": 1, "qualify": "yes"}
label = ["k"]
df2 = pd.DataFrame(dict2, index = label)
df2
new_df = pd.concat([df,df2])
new_df
```
