```python
#method 1
def calculator():
    while True:
        print("what operation do you want?")
        print("+ for addition")
        print("- for subtraction")
        print("* for multiplication")
        print("/ for divition")
        print("exit")
        num1 = float(input("1st number: "))
        num2 = float(input("2nd number: "))
        option = input("operation: ")
        
        if option == "exit":
            print("thanks for using our app!")
            break
            
        if option == "+":
              print(num1 + num2)
        elif option == "-":
              print(num1 - num2)
        elif option == "*":
            print(num1 * num2)
        elif option == "/":
            if num2 != 0:
                print(num1/num2)
            else:
                print("you cant divide by zero")
        else:
            print("invalid operation")
            
         
              
        
```

```python
#Method 2
class calculator:
    def __init__(self, num1=None, num2=None):
        self.num1 = num1
        self.num2 = num2
        
    def add(self, num1, num2):
        print(num1 + num2)
        
    def sub(self, num1, num2):
        print(num1 - num2)
        
    def div(self, num1, num2):
        if num2 != 0:
            print(num1 / num2)
        else:
            print("can't divide by 0")
    def mult(self, num1, num2):
        print(num1*num2)
    
    def menu(self):
        print("what operation do you want?")
        print("+ for addition")
        print("- for subtraction")
        print("* for multiplication")
        print("/ for divition")
        print("exit")
    
    def run(self):
        while True:
            self.menu()
            num1 = float(input("1st number: "))
            num2 = float(input("2nd number: "))
            option = input("operation: ")
            
            
            if option == "exit":
                print("thanks for using our app!")
                break
            
            if option == "+":
                  self.add(num1, num2)
            elif option == "-":
                  self.sub(num1, num2)
            elif option == "*":
                  self.mult(num1, num2)
            elif option == "/":
                self.div(num1, num2)
            else:
                print("invalid operation")
        
       
        
```

```python
app = calculator()
app.run()
```
